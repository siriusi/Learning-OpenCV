```python
import cv2
import numpy as np

img1 = cv2.imread('mavic air.jpg')
img2 = cv2.imread('logo1.jpg')

# 创建ROI区域
rows1,cols1,channels1 = img1.shape
rows2,cols2,channels2 = img2.shape

#roi = img1[rows1-rows2:rows1, cols1-cols2:cols1 ]
img1_center = [int(rows1/2),int(cols1/2)]
img2_center = [int(rows2/2),int(cols2/2)]
roi = img1[(img1_center[0] - img2_center[0]) : (img1_center[0] + img2_center[0]), 
           (img1_center[1] - img2_center[1]) : (img1_center[1] + img2_center[1])]


# 创建Logo的掩膜
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
print(img2gray.shape)
print(img2gray[175,157])
#THRESH_BINARY模式超过170的取最大值255，其余取0
ret, mask = cv2.threshold(img2gray, 170, 255, cv2.THRESH_BINARY) 
cv2.imshow('mask',mask)
mask_inv = cv2.bitwise_not(mask)
cv2.namedWindow('mask_inv')
cv2.imshow('mask_inv', mask_inv)

# 在Mavic Air的ROI区域放上掩膜 即贴上"胶带"
img1_bg = cv2.bitwise_and(roi,roi,mask = mask)
cv2.imshow('img1_bg',img1_bg)

# 现在"刷上"颜色
img2_fg = cv2.bitwise_and(img2,img2,mask = mask_inv)
cv2.imshow('img2_fg',img2_fg)

# "撕掉胶带"
dst = cv2.add(img1_bg, img2_fg)
cv2.imwrite('merge.jpg', dst)
cv2.imshow('dst',dst)
roi[:] = dst

cv2.imshow('res',img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def blurTrackbar(x):
   pass

def GaussianBlurTarckbar(x):
   pass

def boxFilterTarckbar(x):
   pass

img = cv2.imread('1920x1080.jpg')

# 为方便显示 我们将窗口适当缩小
cv2.namedWindow('Blur',cv2.WINDOW_NORMAL)
cv2.resizeWindow('Blur', 480, 640)
cv2.namedWindow('Gaussian Blur',cv2.WINDOW_NORMAL)
cv2.resizeWindow('Gaussian Blur', 480, 640)
cv2.namedWindow('BoxFilter',cv2.WINDOW_NORMAL)
cv2.resizeWindow('BoxFilter', 480, 640)

#cv2.createTrackbar('ksize', 'Blur', 1, 30, blurTrackbar)
cv2.createTrackbar('ksize', 'Gaussian Blur', 1, 30, GaussianBlurTarckbar)
#cv2.createTrackbar('ksize', 'BoxFilter', 1, 30, boxFilterTarckbar)

while(True):
   blurKsize = cv2.getTrackbarPos('ksize', 'Blur')
   GaussianKsize = cv2.getTrackbarPos('ksize', 'Gaussian Blur')
   BoxFilterKsize = cv2.getTrackbarPos('ksize', 'BoxFilter')

#   采用长宽相同的核
   blurImg = cv2.blur(img, (GaussianKsize, GaussianKsize))
#   高斯滤波中 核的尺寸需为正奇数
   if GaussianKsize % 2 is 0:
      GaussianKsize = GaussianKsize + 1
#   采用长宽相同的核
   gauBlurImg = cv2.GaussianBlur(img, (GaussianKsize, GaussianKsize), 0, None, 0)

   boxFilterImg = cv2.boxFilter(img, 1000,(GaussianKsize,GaussianKsize))
    
   cv2.imshow('Blur', blurImg)
   cv2.imshow('Gaussian Blur', gauBlurImg)
   cv2.imshow('BoxFilter', boxFilterImg)
   
   if cv2.waitKey(1) & 0xFF is ord('q'):
      break

cv2.destroyAllWindows()
```

```python
#中值滤波
import cv2
import numpy as np

def medianBlurTrackbar(x):
   pass

img = cv2.imread('1920x1080.jpg')

# 为方便显示 我们将窗口适当缩小
cv2.namedWindow('Median Blur', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Median Blur', 480, 640)
# 为方便显示 我们将窗口适当缩小
cv2.namedWindow('Bilateral Blur', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Bilateral Blur', 480, 640)

cv2.createTrackbar('ksize', 'Median Blur', 1, 30, medianBlurTrackbar)

while (True):
   MedianKsize = cv2.getTrackbarPos('ksize', 'Median Blur')
#  均值滤波中 ksize为正奇数
   if MedianKsize % 2 is 0:
      MedianKsize = MedianKsize + 1

   MedianBlurImg = cv2.medianBlur(img, MedianKsize)
   BilateralBlurImg = cv2.bilateralFilter(img,MedianKsize , MedianKsize*2, int((MedianKsize - 1)/2)) 
   #BilateralBlurImg = cv2.bilateralFilter(img,25 , 25*2, 25/2)
    
   cv2.imshow('Median Blur', MedianBlurImg)
   cv2.imshow('Bilateral Blur', BilateralBlurImg)
    
   if cv2.waitKey(1000) & 0xFF is ord('q'):
      break

cv2.destroyAllWindows()
```
