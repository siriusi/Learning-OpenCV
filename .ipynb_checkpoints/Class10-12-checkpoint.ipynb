```python
#canny算法求边缘
import cv2
import numpy as np

img = cv2.imread('1920x1080.JPG')
# 首先进行高斯滤波 滤除干扰
blurImg = cv2.GaussianBlur(img, (9, 9), 0)

cannyImgGray = cv2.Canny(blurImg, 50, 130)

mask_inv = cv2.bitwise_not(cannyImgGray)

# 现在"刷上"颜色
img2_fg = cv2.bitwise_and(img,img,mask = mask_inv)
cv2.imshow('img2_fg',img2_fg)

cannyImgColor = cv2.cvtColor(cannyImgGray,cv2.COLOR_GRAY2BGR)

b,g,r = cv2.split(cannyImgColor)
r[r>255]=255
g[:]=0
b[:]=0
cannyImgColor = cv2.merge((b, g, r))

# "撕掉胶带"
AndImg = cv2.add(cannyImgColor, img2_fg)

cv2.namedWindow('Canny')
cv2.namedWindow('AndImg')
cv2.imshow('Canny', cannyImgGray)
cv2.imshow('AndImg', AndImg)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#canny算法用在视频上
import numpy as np
import cv2

# 初始化摄像头
cap = cv2.VideoCapture(0)
index = 0
imgname = 0
# 用循环不断获取当前帧 处理后显示出来
while True:
    #   捕获当前帧
    ret,img = cap.read()
    
    blurImg = cv2.GaussianBlur(img, (9, 9), 0)    
    cannyImgGray = cv2.Canny(blurImg, 50, 130)
    mask_inv = cv2.bitwise_not(cannyImgGray)
    #背景
    img_bg = cv2.bitwise_and(img,img,mask = mask_inv)

    cannyImgColor = cv2.cvtColor(cannyImgGray,cv2.COLOR_GRAY2BGR)

    #前景，即边缘
    b,g,r = cv2.split(cannyImgColor)
    g[:]=0
    b[:]=0
    img_fg = cv2.merge((b, g, r))

    # "把前景和背景合成"
    AndImg = cv2.add(img_fg, img_bg)
    cv2.imshow('New Image',AndImg)
    if cv2.waitKey(50) & 0xFF == ord('q'):
        break
# 释放资源
cap.release()
cv2.destroyAllWindows()
```

```python
#膨胀及腐蚀，开操作，闭操作
#dst=cv2.dilate(src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]) 膨胀
#dst=cv2.erode(src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]) 腐蚀
#dst=cv2.morphologyEx(src, op, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]]) 形态学操作，op取值见上
import cv2
import numpy as np

img = cv2.imread('testPhoto.JPG')
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (8, 8))
dilateImg = cv2.dilate(img, kernel)
erodeImg = cv2.erode(img, kernel)

openImg = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closeImg = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradImg = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

cv2.imshow('SrcImg', img)
cv2.imshow('DilateImg', dilateImg)
cv2.imshow('ErodeImg', erodeImg)

cv2.imshow('openImg', openImg)
cv2.imshow('closeImg', closeImg)
cv2.imshow('gradImg', gradImg)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#形态学开操作不同核心比较 开操作是先腐蚀后膨胀，所以较小的点被腐蚀后没掉了，再膨胀其他较大的点回复了，而小点就没掉了
#对比闭操作，闭操作是先膨胀后腐蚀，膨胀后很多之前的间隙被填上，再腐蚀这些间隙也不会再出现了
import cv2
import numpy as np

img = cv2.imread('cells.jpg')
cellGrayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, cellBinImg = cv2.threshold(cellGrayImg, 150, 255, cv2.THRESH_BINARY)

kernel8 = cv2.getStructuringElement(cv2.MORPH_RECT, (8, 8))
openImg8 = cv2.morphologyEx(cellBinImg, cv2.MORPH_OPEN, kernel8)

kernel4 = cv2.getStructuringElement(cv2.MORPH_RECT, (4, 4))
openImg4 = cv2.morphologyEx(cellBinImg, cv2.MORPH_OPEN, kernel4)

cv2.imshow('BinImg', cellBinImg)
cv2.imshow('OpenImg4', openImg4)
cv2.imshow('OpenImg8', openImg8)

closeImg4 = cv2.morphologyEx(cellBinImg, cv2.MORPH_CLOSE, kernel4)
closeImg8 = cv2.morphologyEx(cellBinImg, cv2.MORPH_CLOSE, kernel8)
cv2.imshow('closeImg4', closeImg4)
cv2.imshow('closeImg8', closeImg8)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#使用开操作和闭操作进行滤波处理
import cv2
import numpy as np

img = cv2.imread('./dip_images_09/Fig0911(a)(noisy-fingerprint).jpg')
cellGrayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cellGrayImg[cellGrayImg > 150] = 255
cellGrayImg[cellGrayImg < 150] = 0
cv2.imshow('Source', cellGrayImg)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

tempImg = cv2.morphologyEx(cellGrayImg, cv2.MORPH_ERODE, kernel)    #腐蚀
cv2.imshow('ERODE', tempImg)

tempImg = cv2.morphologyEx(tempImg, cv2.MORPH_DILATE, kernel)      #腐蚀+膨胀=开操作
cv2.imshow('OPEN', tempImg)

tempImg = cv2.morphologyEx(tempImg, cv2.MORPH_DILATE, kernel)      #开操作+膨胀
cv2.imshow('OPEN & DILATE', tempImg)

tempImg = cv2.morphologyEx(tempImg, cv2.MORPH_ERODE, kernel)       #开操作+闭操作
cv2.imshow('OPEN & CLOSE', tempImg)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#霍夫算法检测直线
import cv2
import numpy as np

def drawfindLines_Hough(img, lines):
    if lines is None:
        return
    for i in lines:
        for rho, theta in i:
            a = np.cos(theta)
            b = np.sin(theta)
#           变换到x-y坐标系
            x0 = a * rho
            y0 = b * rho
#           用点和斜率得到直线上的两个点
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 1)

img = cv2.imread('./dip_images_09/Fig0911(a)(noisy-fingerprint).jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 100, apertureSize = 3)
cv2.imshow('edges', edges)

lines = cv2.HoughLines(edges, 1, np.pi/180,100)
drawfindLines_Hough(img, lines)

cv2.imshow('Line', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#霍夫算法检测圆
#circles=cv2.HoughCircles(image, method, dp, minDist[, circles[, param1[, param2[, minRadius[, maxRadius]]]]])
import cv2
import numpy as np

img = cv2.imread('duolaAmeng.jpg')
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

circle = cv2.HoughCircles(grayImg, cv2.HOUGH_GRADIENT, 1, 50, minRadius = 150, maxRadius = 200)

for c in circle:
   for x,y,r in c:
      cv2.circle(img, (x,y), r, (0, 0, 255),10)

cv2.imshow('Circle', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#用矩形圈出轮廓
import cv2
import numpy as np

img = cv2.imread('duolaAmeng.jpg')
# 图像预处理
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, binImg = cv2.threshold(grayImg, 100, 255, cv2.THRESH_BINARY)
# 寻找轮廓
_, contours, hierarchy = cv2.findContours(binImg, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print("contours number: ", len(contours)) #总轮廓数量
# 画出轮廓 image, contours, contourIdx, color 其中contours = 1的轮廓为当前显示的轮廓
# 由于是树形轮廓，contours = 0是整个img的外圈
cv2.drawContours(img, contours, 1, (0, 200, 0), 2)
# CHAIN_APPROX_SIMPLE逼近模式下该轮廓有798个点 若逼近模式为CHAIN_APPROX_NONE则有1609个点
print(contours[1].shape)
# 用方框框出轮廓
x, y, w, h = cv2.boundingRect(contours[1])
cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 200), 2)
cv2.imshow('Contours Image', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#视频轮廓
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
while True:
    if cap.isOpened():
        ret,img = cap.read()
        #cv2.imshow('Camera',frame)
        gauBlurImg = cv2.GaussianBlur(img, (9,9), 0, None, 0)
        #img = cv2.imread('duolaAmeng.jpg')
        # 图像预处理
        grayImg = cv2.cvtColor(gauBlurImg, cv2.COLOR_BGR2GRAY)
        
       
        
        _, binImg = cv2.threshold(grayImg, 0, 255, cv2.THRESH_OTSU)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (8, 8))
        erodeImg = cv2.erode(binImg, kernel, iterations = 10)
        dilateImg = cv2.dilate(binImg, kernel, iterations = 10)
        
        
        # 寻找轮廓
        _, contours, hierarchy = cv2.findContours(binImg, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        cv2.drawContours(img, contours, -1, (0, 200, 0), 2)
        if len(contours) > 0:
            # 用方框框出轮廓
            x, y, w, h = cv2.boundingRect(contours[0])
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 200), 2)
        cv2.imshow('Contours Image', img)
        
        if cv2.waitKey(100)&0xFF is ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
