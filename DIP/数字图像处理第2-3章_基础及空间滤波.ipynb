```python
#拆分成8个bit图层，每个图层是8bit灰度值中的1bit
import cv2
import numpy as np

img = cv2.imread('../images/1920x1080.JPG',1)
img2gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print(np.max(img2gray))
cv2.namedWindow('Source Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Source Image', 640, 480)
cv2.imshow('Source Image', img2gray)

allOnes = np.ones_like(img2gray)

bitMash = 1

for i in range(8):
    bitMash = 1 << i
    mashMat = bitMash * allOnes
    #print(mashMat)
    #print(img2gray)
    #print()
    newImg = cv2.bitwise_and(mashMat,img2gray)
    imgName = 'Bit ' + str(i) + " Image"
    cv2.namedWindow(imgName, cv2.WINDOW_NORMAL)
    cv2.resizeWindow(imgName, 640, 480)
    cv2.imshow(imgName, newImg)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#局部增强
import cv2
import numpy as np


def getNeighbourhood(img, x, y, dim = 3):
    if dim % 2 == 0:
        return
    xAsix = range(x - int(dim / 2), x + int(dim / 2) + 1)
    yAsix = range(y - int(dim / 2), y + int(dim / 2) + 1)
    #print(xAsix)
    #print(yAsix)
    lenOfX = img.shape[1]
    lenOfY = img.shape[0]
    neighbourhoodGray = []
    for i in xAsix:
        for j in yAsix:
            if i >= 0 and i < lenOfX and j >= 0 and j < lenOfY:
                neighbourhoodGray.append(img[j,i])
    return neighbourhoodGray

img = cv2.imread('../images/1920x1080.JPG',1)
img2gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cv2.namedWindow('Source Image', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Source Image', 800, 450)
cv2.imshow('Source Image', img2gray)

[rows,cols] = img2gray.shape
mean_G, stddev_G = cv2.meanStdDev(img2gray)
newImg = img2gray
for x in range(cols):
    for y in range(rows):
        mean, stddev = cv2.meanStdDev(np.array(getNeighbourhood(img2gray,x,y)))
        if mean <= 0.8 * mean_G and stddev >= 0.02 * stddev_G and  stddev <= 0.4 * stddev_G:
            newImg[y,x] = np.min([255, 4 * img2gray[y,x]])

cv2.namedWindow('NewImage', cv2.WINDOW_NORMAL)
cv2.resizeWindow('NewImage', 800, 450)
cv2.imshow('NewImage', newImg)

cv2.waitKey(0)
cv2.destroyAllWindows()


```
